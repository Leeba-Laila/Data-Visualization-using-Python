### 1. Plot a pie chart for columns: ‘cyl’ and ‘model’ form the mtcars.csv data frame

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv(r"C:\Users\91730\Desktop\Python_Intellipaat\Python\Datasets\mtcars.csv")
df.head()
```

```python
plt.figure(figsize=(8,8))
plt.pie(df['cyl'], labels=list(df['model']))
plt.title("Pie chart representation of models")
plt.show() 
```
