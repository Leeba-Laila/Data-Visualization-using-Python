```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (3,3))
x = (40, 20, 5)
label = ['Bikes', 'Cars', 'Bus']
plt.pie(x, labels = label, autopct = '%1.1f%%')
plt.show()
```

```python
plt.figure(figsize=(6,6))
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23, 17, 35, 29, 12]
cols = ['c', 'm', 'r', 'b', 'g']
plt.pie(students, labels = langs, colors = cols, startangle = 90, shadow = False, explode = (0,0,0,0,0.2), autopct = '%1.1f%%')
plt.show()
```

```python
#Error Bar

plt.style.use('seaborn-whitegrid')
x = np.arange(10)/10
y = (x +0.1) **2   #(a+b)**2
yerror = np.linspace(0.05, 0.2, 10)
plt.errorbar(x, y, yerr = yerror, fmt = 'or')
plt.show()
```

```python
# fmt = '-og'

plt.style.use('classic')
x = np.arange(10)/10
y = (x +0.1) **2
yerror = np.linspace(0.05, 0.2, 10)
plt.errorbar(x, y, yerr = yerror, fmt = '-og')
plt.show()
```

```python
plt.style.use('seaborn-whitegrid')
x = np.arange(10)/10
y = (x +0.1) **2   #(a+b)**2
yerror = np.linspace(0.05, 0.2, 10)
plt.errorbar(x, y, yerr = yerror, fmt = '-ob')
plt.xlabel = 'X-axis'
plt.ylabel = 'Y-axis'
plt.title("Line plot with error bars")
plt.show()
```

```python
x = np.arange(10)/10
y = x **2
xerror = 0.1
plt.errorbar(x, y, xerr = xerror, fmt = '-oy')
plt.xlabel = 'X-axis'
plt.ylabel = 'Y-axis'
plt.title = 'Horizontal error bar'
plt.show()
```
