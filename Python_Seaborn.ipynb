### Boxplot
* Box plot conveys the distribution of values, the maximum and median values.
* Box plot is used for a numeric variable (univariate Analysis)
* Draw a box plot to show distributions with respect to categories.(Bivariate Analysis)

### Box plot and Whiskers
*The Concept of median, percentile, quantile, whiskers
*A box and whisker plotâ€”also called a box plotâ€”displays the five-number summary of a set of data.
*The five-number summary is the minimum, first quartile, median, third quartile, and maximum. In a box plot
*Box plots may also have lines extending from the boxes indicating variability outside the upper and lower quartiles

```python
#Univariate Analysis with a boxplot

sns.boxplot(x = 'total_bill', data = tips_df)
plt.show()
```

```python
#Univariate Analysis with a boxplot

sns.boxplot(x = 'tip', data = tips_df)
plt.show()
```

```python
#Univariate Analysis with a boxplot

sns.boxplot(y = 'tip', data = tips_df)
plt.show()
```

- By default, the x axis is considered as the categorical data and y as numerical data.
- If two numerical values are to be plotted, we have to mention the discrete value on x axis and continuos value on y axis

```python
#Bivariate Analysis with a boxplot(Continuous vs Categorical)
#Relationship between categorical and continuous variable 

sns.boxplot(x = 'tip', y = 'day', data = tips_df)
plt.show()
```

```python
sns.boxplot(x = 'tip', y = 'day', data = tips_df, palette='spring')
plt.show()
```

- If both variables are numeric and we don't specify orient, by default it takes x axis as categorical and y as continuous

```python
#Here we cannot get a proper plot since both the x and y values are numerical data.
#Boxplot works only for the combination of categorical and numerical data

sns.boxplot(x = 'tip', y = 'size', data = tips_df)
plt.show()
```

```python
#Here when the orient is given as 'h', we mention that the median has to be calculated on the tip and not on size.
#So even if specify  x as tip and y as size since we specify the orient, it would calculate median for tip.

sns.boxplot(x='tip', y='size', data=tips_df, palette='spring', orient='h')
plt.show()
```

### Joint Plot :
* The joint plot allows us to draw a plot of two variables with bivariate and univariate graphs.
* Draw multiple bivariate plots with univariate marginal distributions.
* We can also specify the kind of plots with the â€˜kindâ€™ keyword argument.
* Kind must be either â€˜scatterâ€™, â€˜regâ€™, â€˜residâ€™, or â€˜hexâ€™.

```python
#In this case, we get a view of univariate analysis(barplot) and Bi-variate analysis(scatterplot)

sns.jointplot(x = 'total_bill', y = 'tip', data = tips_df, kind = 'scatter')
plt.show()
```

- Residue isthe difference between the the actual value and the value from the line

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips_df, kind = 'reg')
plt.show()

#According to my regression line, the tip for $50 is somewhere near $6. But when I consider the scatterplot, 
#the tip for $50 is showing somewhere near to $10. This difference between the actual data(ie, data from scatterplot 
#which is 10) and the data from the line(ie, data from the regression line which is $6) is called the residue. This shows 
#the error margin
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips_df, kind = 'resid')
plt.show()
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips_df, kind = 'hex')
plt.show()
```

# Correlation with heatmap

- Heatmap helps us to find and visualize correlation between each and every feature
- Correlation values lies between -1 and 1
- if -1, the data are negatively correlated
- if 1, the data are positively correlated
- if value more than 0.5, then the data is highly correlated. If less than 0.5, the data is less correlated.

```python
print(tips_df.corr())
```

```python
sns.heatmap(tips_df.corr())
plt.show()
```

```python
#annot helps to print the data on top of the map in each box

sns.heatmap(tips_df.corr(), annot = True, cmap = 'RdBu')
plt.show()
```

### Violin plot
* It is a combination of boxplot and kernel density distribution
* Show the relationship between a nuerical and categorical data

```python
#Shows the relationship between total bill and day

sns.violinplot(x = 'total_bill', y = 'day', data = tips_df)
plt.show()

# You have a boxplot inside the violinplot. The data beyond the whiskers of the boxplot inside the violinplot specifies 
#the outliers. This also shows how the data is distributed. Also shows the min ans max values. Therefore violinplot 
#is also used in the industry.
```

```python
#Shows the relationship between tip and day

sns.violinplot(x = 'tip', y = 'day', data = tips_df)
plt.show()
```

```python
#Shows the relationship between size and total bill

sns.violinplot(x = 'size', y = 'total_bill', data = tips_df)
plt.show()
```

```python
#Shows the relationship between sex and total bill

sns.violinplot(x = 'sex', y = 'total_bill', data = tips_df)
plt.show()
```
